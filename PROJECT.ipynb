```python
# Variabel independen (X) dan dependen (y)
X = data[['TINGKAT PENGANGGURAN (%)', 'PENDUDUK MISKIN (RIBU)']]
y = data['TINGKAT PERCERAIAN']
```

```python
# Menambahkan konstanta untuk model regresi
X = sm.add_constant(X)

# Membuat model regresi
model = sm.OLS(y, X)

# Melatih model
hasil = model.fit()

# Menampilkan summary hasil regresi
print(hasil.summary())
```

Dengan menggunakan OLS dapat memberikan informasi tentang hubungan antara tingkat perceraian (variabel dependen) dengan tingkat pengangguran dan jumlah penduduk miskin di suatu daerah atau wilayah. Adapun beberapa informasi yang diperoleh dari hasil ini :
- Nilai R-squared sebesar 0.617 menunjukkan bahwa model dapat menjelaskan sekitar 61.7% variasi dari tingkat perceraian yang diamati.
- Adjusted R-squared (Adj. R-squared) menyesuaikan R-squared untuk jumlah variabel independen dalam model. Nilai Adjusted R-squared sebesar 0.585 menunjukkan bahwa sekitar 58.5% variasi dalam tingkat perceraian dapat dijelaskan oleh variabel-variabel yang dimasukkan dalam model.
- F-statistic (F-ratio) digunakan untuk menguji signifikansi keseluruhan dari model regresi. Nilai F-statistic sebesar 19.32 dengan Prob (F-statistic) sebesar 9.99e-06 (atau 0.00000999) menunjukkan bahwa model secara keseluruhan signifikan secara statistik. Model ini dapat menjelaskan sebagian besar variasi dalam tingkat perceraian, meskipun terdapat beberapa variabel lain yang mungkin perlu dimasukkan untuk meningkatkan keakuratan dan keberlakuannya.
- Intercept (const): Koefisien intercept (konstanta) adalah 1730.4727 dengan standard error (std err) sebesar 1115.605. P-value (P>|t|) untuk intercept adalah 0.134, yang berarti tidak signifikan secara statistik pada tingkat signifikansi 0.05. Ini menunjukkan bahwa ketika variabel independen (TINGKAT PENGANGGURAN (%) dan PENDUDUK MISKIN (RIBU)) adalah nol, tingkat perceraian di wilayah tersebut diperkirakan berada pada nilai 1730.4727, meskipun tidak signifikan.
- TINGKAT PENGANGGURAN (%): Koefisien untuk variabel ini adalah -88.1942 dengan std err sebesar 145.864. P-value untuk TINGKAT PENGANGGURAN (%) adalah 0.551, yang berarti tidak signifikan secara statistik pada tingkat signifikansi 0.05. Ini menunjukkan bahwa tidak ada bukti yang cukup kuat untuk mendukung hubungan antara tingkat pengangguran dengan tingkat perceraian di wilayah tersebut berdasarkan model ini.
- PENDUDUK MISKIN (RIBU): Koefisien untuk variabel ini adalah 0.0187 dengan std err sebesar 0.003. P-value untuk PENDUDUK MISKIN (RIBU) adalah 0.000, yang menunjukkan bahwa variabel ini signifikan secara statistik pada tingkat signifikansi 0.05. Ini berarti bahwa penambahan satu unit dalam jumlah penduduk miskin (dalam ribu) dikaitkan dengan peningkatan sebesar 0.0187 unit dalam tingkat perceraian.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Pilih variabel untuk clustering
X = data[['NIKAH', 'CERAI TALAK', 'CERAI GUGAT', 'TINGKAT PENGANGGURAN (%)', 'PENDUDUK MISKIN (RIBU)']]

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Metode Elbow untuk menentukan jumlah klaster optimal
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plotting Elbow Method
plt.plot(range(1, 10), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python

# Menggunakan algoritma K-means untuk clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
data['Cluster'] = kmeans.labels_

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', s=50, alpha=0.8)
plt.title('Clustering Kabupaten/Kota')
plt.xlabel('Skor K-means 1')
plt.ylabel('Skor K-means 2')
plt.colorbar(label='Cluster')
plt.show()

# Menampilkan hasil clustering
print(data[['KABUPATEN/KOTA', 'Cluster']])
```

Korelasi yang tinggi antara tingkat perceraian dan jumlah penduduk miskin (ribu) menunjukkan bahwa faktor ini mungkin memiliki pengaruh yang signifikan terhadap tingkat perceraian di daerah-daerah tersebut.
Sebaliknya, tingkat pengangguran dalam bentuk persentase tidak menunjukkan hubungan yang signifikan dengan tingkat perceraian berdasarkan korelasi tersebut.

Dalam penelitian ini, menggunakan 3 cluster dalam mengelompokkan daerah kabupaten dan kota yang memiliki karakteristik yang sama . berikut pembagian dari ketiga cluster tersebut:
- Cluster 0: Kabupaten/kota dalam klaster ini memiliki karakteristik yang relatif rendah dalam hal jumlah perkawinan, perceraian, tingkat pengangguran, dan penduduk miskin dibandingkan dengan klaster lainnya. Secara umum, ini mungkin menunjukkan kabupaten/kota dengan profil sosial-ekonomi yang stabil atau lebih rendah dalam hal perubahan status pernikahan. Daerah yang termasuk dalam cluster ini yaitu Sukabumi, Cianjur, Tasikmalaya, Ciamis, Kuningan, Majalengka, Sumedang, Subang, Karawang, Bekasi, Bandung Barat, Kota Bandung, Kota Bekasi dan Kota Depok.

- Cluster 1: Kabupaten/kota dalam klaster ini cenderung memiliki tingkat perceraian yang lebih tinggi dan mungkin juga tingkat pengangguran dan jumlah penduduk miskin yang lebih tinggi.
```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Data frame Anda
# Misalnya, X adalah variabel independen dan y adalah variabel dependen
X = data[['TINGKAT PENGANGGURAN (%)', 'PENDUDUK MISKIN (RIBU)']]
y = data['TINGKAT PERCERAIAN']

# Bagi data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inisialisasi model regresi linier
model = LinearRegression()

# Latih model menggunakan data training
model.fit(X_train, y_train)

# Prediksi menggunakan data testing
y_pred = model.predict(X_test)

# Evaluasi performa model (misalnya, dengan menggunakan Mean Squared Error)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

Nilai MSE menunjukkan bahwa rata-rata kuadrat dari selisih antara prediksi model dan nilai yang diamati adalah sebesar 3,454,590.1913405447 .
